```python
import pandas as pd
import glob
import re
```

```python
files = glob.glob('*.txt')
files
```

```python
for file in files:
    with open(file, 'r') as fp:
        lines = fp.readlines()
        lines[:10]
    cat_re = re.compile("(.*) (.{1,5})")

    output = 'Categorie,Rang,Nom,Naissance,Lieu,Groupe,Temps,Dossard\n'
    for line in lines[4:-4]:
        if line.find(',') < 0:
            if len(line) > 1:
                category = re.match(cat_re, line).group(1)
                print("{}".format(category))
        else:
            output += '"{}",{}'.format(category,line)

    with open('New_'+file, 'w') as outputfile:
        outputfile.write(output)
```

```python
df_files = glob.glob('New_*.txt')
df_files
```

```python
df_file = df_files[0]
df_file[-8:-4]
```

```python
def rank_as_int(value):
    try:
        return int(value[:-1])
    except ValueError as e:
        return -1
        print(value, e)
```

```python
whole_df = pd.DataFrame()
for df_file in df_files:
    df = pd.read_csv(df_file)
    df['Rang'] = df['Rang'].apply(rank_as_int)
    df['year'] = df_file[-8:-4]
    df = df[df['Rang'] != -1]
    whole_df = whole_df.append(df)
```

```python
whole_df.to_csv('whole_df_datasport.csv', index=False)
```
